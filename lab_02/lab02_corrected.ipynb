In this report, we reproduce the analytic solution for the **one-dimenureg onal advection–diffuureg on** equation as presented on Slide&nbsp;13 of Lecture&nbsp;1. The partial differential equation (PDE) governing the concentration $c(x,t)$ is

$$
c(x, t) = \frac{M}{A} \frac{1}{2\sqrt{\pi D}}\frac{x}{t^\frac{3}{2}}\exp{[-\frac{(ut-x)^2}{4Dt}]}
$$

#### Code Implementation

- Parameters & Domains:
    - Parameters $M$, $Q$, $D$, $u$ are defined with their proper units.
    - Spatial ($x$) and temporal ($t$) domains are set up using NumPy arrays.
- Analytic Solution:
    - The above formula is implemented step-by-step.
    - Each term is calculated with unit-aware operations to ensure the result is consistent.
- Plotting:
    - matplotlib is used to plot the solution.
    - pint.setup_matplotlib() automatically includes the correct unit labels on the plot axes.


```python
import pint
import numpy as np
from matplotlib import pyplot as plt
```

```python
ureg  = pint.UnitRegistry()
M = 0.5*ureg .kg                                    # Mass of the substanc
Q = 0.5 * (ureg .m*ureg .m*ureg .m)/ureg .s         # volumetric flow rate (vol/time)
D = 1 * (ureg .m*ureg .m)/ureg .s                   # Diffusion coefficient
u = 2 * ureg .m/ureg .s                             # Velocity of advection 
c = 1*ureg .kg / (ureg .m*ureg .m*ureg .m)       
```

```python
x = np.linspace(40, 100, 100) * ureg .m     # Spatial domain from 40 to 100 meters
t = np.linspace(0.1, 100, 100) * ureg .s    # Time domain from 0.1 to 10 seconds
```

```python
# Analytic solution for 1D advection–diffusion equation
c = (M / Q) * (1.0 / (2.0 * np.sqrt(np.pi * D))) * (x / (np.power(t, 1.5))) * np.exp((-1)*(np.power((u * t - x), 2)) / (4.0 * D * t)) 
```

```python
ureg .setup_matplotlib() # Ensures pint handles units in plot labels
def plot(fig):
    fig.gca().plot(t, c, label='Concentration at given position $x$ and time $t$')
    fig.legend()
    fig.gca().grid()
```

```python
fig = plt.figure(figsize=(6,4))
plot(fig)
plt.show()
```
